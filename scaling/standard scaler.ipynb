```python
import numpy as np
import pandas as pd
```

Normalization is the process of rescaling one or more attributes to the range of 0 to 1. This means that the largest value for each attribute is 1 and the smallest value is 0

X' = (X-Xmin)/(Xmax-Xmin) where
Xmax = Maximum value
Xmin = Minimum value

Method of Feature Scaling:
    1. Standard Scaler
    2. Minmax Scaler
    3. Robust Scaler
    4. Normalizer

StandardScaler: Assumes your data is normally d=istributed within each feature and will scale them such that the distribution is now centered around 0

```python
from sklearn import preprocessing
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns

```

```python
np.random.seed(1)

df = pd.DataFrame(
    {
        'x1':np.random.normal(0,2,10000),
        'x2':np.random.normal(5,3,10000),
        'x3':np.random.normal(-5,5,10000)
    }
)
```

```python
df
```

Apply Standard Scaler

```python
scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(
    scaled_df,
    columns=['x1','x2','x3']
)
```

```python
scaled_df
```

```python
scaled_df.hist(bins =30)
plt.tight_layout()
```

```python
df.hist(bins = 30)
plt.tight_layout()
```

```python
scaled_df.x1.std()
```

```python
df.x1.std()
```

```python
# plotting before scaling and after scaling data
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize = (10,10))

# before scaling

ax1.set_title('Before scaling')
columns = df.columns
for column in columns:
    sns.kdeplot(df[column], ax=ax1, shade=True)

# after scaling

ax2.set_title('After Scaling')
columns = scaled_df.columns
for column in columns:
    sns.kdeplot(scaled_df[column], ax=ax2, shade=True)

plt.tight_layout()
```
