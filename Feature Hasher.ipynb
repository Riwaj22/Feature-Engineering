```python
import numpy as np
import pandas as pd

```

```python
data = pd.read_csv(r"C:\Users\97798\Desktop\Feature Engineering\Categorical data\Pokemon.csv")
```

```python
data.drop('#', axis=1, inplace=True)
```

```python
data
```

```python
data.Generation  = data.Generation.map(
    {
        1:'Gen 1',
        2:'Gen 2',
        3:'Gen 3',
        4:'Gen 4',
        5:'Gen 5',
        6:'Gen 6'
    }
)
```

```python
data.Generation
```

```python
from sklearn.feature_extraction import FeatureHasher

fh = FeatureHasher(n_features=5, input_type='string')
hashed_features = fh.fit_transform(data['Generation'])
hashed_features = hashed_features.toarray()
new_df = pd.DataFrame(hashed_features)

pd.concat([data, new_df],axis =1)
```
