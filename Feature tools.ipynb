```python
import pandas as pd

# Sample data for demonstration
data = {
    'client_id': range(1, 101),
    'joined': pd.date_range(start='2021-01-01', periods=100),
    'income': [50000, 60000, 45000, 70000, 55000] * 20,  # Repeating pattern for simplicity
    'credit_score': [700, 720, 680, 750, 710] * 20  # Repeating pattern for simplicity
}

# Create DataFrame
clients = pd.DataFrame(data)

# Display the DataFrame
clients.head()
```

```python
clients.set_index('client_id', inplace=True)
```

```python
import pandas as pd

# Sample data for demonstration
data = {
    'payment_id': range(1, 101),
    'loan_id': [i for i in range(1, 101)],
    'payment_amount': [500, 1000, 750, 250, 500] * 20,  # Repeating pattern for simplicity
    'payment_date': pd.date_range(start='2021-01-01', periods=100),
    'missed': [False, False, True, False, True] * 20  # Repeating pattern for simplicity
}

# Create DataFrame
payment = pd.DataFrame(data)

# Display the DataFrame
payment.head()
```

```python
import pandas as pd

# Sample data for demonstration
data = {
    'loan_id': range(1, 101),
    'client_id': [i for i in range(1, 101)],
    'loan_type': ['personal', 'business', 'mortgage', 'personal', 'student'] * 20,  # Repeating pattern for simplicity
    'loan_amount': [10000, 20000, 150000, 5000, 25000] * 20,  # Repeating pattern for simplicity
    'repaid': [True, True, False, True, False] * 20,  # Repeating pattern for simplicity
    'loan_start': pd.date_range(start='2021-01-01', periods=100),
    'loan_end': pd.date_range(start='2021-02-01', periods=100),
    'rate': [0.05, 0.08, 0.06, 0.04, 0.07] * 20  # Repeating pattern for simplicity
}

# Create DataFrame
loan = pd.DataFrame(data)

# Display the DataFrame
loan.head()
```

```python
import featuretools as ft

# Create an empty EntitySet
es = ft.EntitySet(id='financial_data')

# Add clients dataframe as an entity
es = es.add_dataframe(
    dataframe=clients,
    dataframe_name='clients',
    index='client_id',
    time_index='joined'
)

# Add loans dataframe as an entity
es = es.add_dataframe(
    dataframe=loan,
    dataframe_name='loans',
    index='loan_id',
    time_index='loan_start'
)

# Add payments dataframe as an entity
es = es.add_dataframe(
    dataframe=payment,
    dataframe_name='payments',
    index='payment_id',
    time_index='payment_date'
)

# Continue with further feature engineering or entity creation
```

```python
es.add_relationship(
    parent_dataframe_name='loans',
    parent_column_name='loan_id',
    child_dataframe_name='payments',
    child_column_name='loan_id'
)
```
